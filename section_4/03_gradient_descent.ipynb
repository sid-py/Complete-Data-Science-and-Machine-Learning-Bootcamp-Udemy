# Example 1 : A simple cost function

$f(x) = x**2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
# Make Data
x_1 = np.linspace(start = -3, stop=3, num = 500)
x_1
```

```python
plt.plot(x_1, f(x_1))
plt.show()
```

## Slopes and Derivaties

```python
def df(x):
    return 2*x + 1
```

```python
df_x = df(x_1)
df_x
```

```python
plt.plot(x_1, df(x_1))
plt.show()
```

```python
# Gradient Descent

new_x = 3
prev_x = 0
step_multiplier = 0.1
precision = 0.0001
counter = 0
while True:
    prev_x = new_x
    gradient = df(prev_x)
    new_x = prev_x - (step_multiplier * gradient)
    
    step_size = abs(new_x - prev_x)
    counter += 1
    if step_size < precision:
        break
print(f"The loop ran {counter} times.")
    
print("Local minimum occurs at: ", new_x)
print("Slope of df(x) value at this point is: ", df(new_x))
print("f(x) value or cost at this point is:  ", f(new_x))
```

# Example 2 - Multiple minimum v/s Initial Guess and Advanced Functions

## $$g(x) = x ^4 - 4x^2 + 5$$

```python
#  Make some data
x_2 = np.linspace(-2,2,1000)

def g(x):
    return x**4 - 4*x**2 + 5

def dg(x):
    return 4*x**3 - 8*x
```

```python
#  First Plot
plt.subplot(1,2,1)
plt.plot(x_2, g(x_2))
plt.xlabel("x_2")
plt.ylabel("g (x_2)")
        
#  Second Plot
plt.subplot(1,2,2)
plt.plot(x_2, dg(x_2))
plt.xlabel("x_2")
plt.ylabel("dg (x_2)")
plt.show()
```

## Gradient Descent as a Python Function

```python
# Gradient Descent
def gradient_descent(derivative_func, initial_guess, multiplier = 0.02, precision = 0.001, max_iter = 300):
    new_x = initial_guess

   
    x_list = [new_x]
    
    slope_list = [derivative_func(new_x)]
    counter = 0
    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - (multiplier * gradient)
        
        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        counter += 1
        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

```python
local_min, list_x , deriv_list = gradient_descent(dg, 0.5, 0.02, 0.001)
print("Local min occurs at : ", local_min)
print("No. of steps: ", len(list_x))
```

```python
local_min, list_x , deriv_list = gradient_descent(dg, -0.5, 0.01)
print("Local min occurs at : ", local_min)
print("No. of steps: ", len(list_x))
```

```python
#  Calling Gradient Descent function
#  Plotting function and derivative and scatter plot side by side
plt.subplot(1,2,1)
plt.plot(x_2, g(x_2), color= "skyblue", alpha = 0.6)
plt.xlabel("x_2")
plt.ylabel("g (x_2)")

plt.scatter(list_x, g(np.array(list_x)), color = "r", s = 100, alpha = 0.5 )
        
#  Second Plot
plt.subplot(1,2,2)
plt.plot(x_2, dg(x_2), color= "skyblue", alpha = 0.6)
plt.xlabel("x_2")
plt.ylabel("dg (x_2)")
plt.scatter(list_x, deriv_list, color = "r", s = 100, alpha = 0.5 )
plt.show()
```

# Example 3 - Divergence, Overflow and Python Tuples

## $$ h(x) = x^5 - 2x^4 + 2 $$

```python
#  Make Data
x_3 = np.linspace(start = -2.5, stop = 2.5, num = 1000)

def h(x):
    return x**5 - 2*x**4 + 2

def dh(x):
    return 5*x**4 - 8*x**3
```

```python
#  Calling Gradient Descent function

local_min, list_x , deriv_list = gradient_descent(derivative_func= dh, initial_guess= -0.2, max_iter=10)
print("Local min occurs at : ", local_min)
print("No. of steps: ", len(list_x))
#  Plotting function and derivative and scatter plot side by side
plt.subplot(1,2,1)
plt.xlim(-1.2, 2.5)
plt.ylim(-1, 4)
plt.plot(x_3, h(x_3), color= "skyblue", alpha = 0.6)
plt.xlabel("x_2")
plt.ylabel("g (x_2)")

plt.scatter(list_x, h(np.array(list_x)), color = "r", s = 100, alpha = 0.5 )
        
#  Second Plot
plt.subplot(1,2,2)
plt.xlim(-1, 2)
plt.ylim(-4, 5)
plt.plot(x_3, dh(x_3), color= "skyblue", alpha = 0.6)
plt.xlabel("x_2")
plt.ylabel("dg (x_2)")
plt.scatter(list_x, deriv_list, color = "r", s = 100, alpha = 0.5 )
plt.show()

print("LOcal min occurs at:", local_min)
print("Cost at this Minimum is:", h(local_min))
print()
```
