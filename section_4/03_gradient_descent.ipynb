# Example 1 : A simple cost function

$f(x) = x**2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
# Make Data
x_1 = np.linspace(start = -3, stop=3, num = 500)
x_1
```

```python
plt.plot(x_1, f(x_1))
plt.show()
```

## Slopes and Derivaties

```python
def df(x):
    return np.diff(x, n=0)
```

```python
df_x = df(x_1)
df_x
```

```python
plt.plot(x_1, df(x_1))
plt.show()
```
